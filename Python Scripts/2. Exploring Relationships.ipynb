### Create a correlation heatmap using matplotlib

```python
plt.matshow(df_sub.corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap
f = plt.figure(figsize=(10, 10)) # figure size 
plt.matshow(df.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_sub.shape[1]), df_sub.columns, fontsize=10, rotation=80) # x axis labels
plt.yticks(range(df_sub.shape[1]), df_sub.columns, fontsize=10) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=11) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

## 6. Creating a Categorical Plot 

```python
# Use a histogram to visualize the distribution of the variables. 
# This way, you can determine sensible categories for the price ranges. 
# You don't want to end up with too few observations in any one of the categories.
# The argument "kde" add a line that encompasses the distribution

sns.histplot(df['Happiness Score'], bins = 20, kde = True)
```

```python
# Creating a Happiness flag for Countries
df.loc[df['Happiness Score'] < 4, 'Happiness Status'] = 'Unhappy Country'
df.loc[(df['Happiness Score'] >= 4) & (df['Happiness Score'] < 6), 'Happiness Status'] = 'Somewhat Happy Country'
df.loc[df['Happiness Score'] >= 6, 'Happiness Status'] = 'Happy Country'
```

```python
# Checking creation of Exclusion flag 

df['Happiness Status'].value_counts(dropna = False)
```

```python
df.columns
```

```python
# Creating a categorical plot in seaborn using the Happiness Status created above

sns.set(style="ticks")
g = sns.catplot(x="Country", y="Happiness Score", hue="Happiness Status", data=df)
```

This categorical plot displays that a majority of the countries in this study land within the "Somewhat Happy Country" status. We see a few of countries in the "Unhappy Country" status and the rest of the countries landing within the "Happy Country status.
